```python
import matplotlib.pyplot as plt
```

```python
printer_data = {
    'name': 'Printer 1',
    'ip_address': '123.455.433.001',
    'cyan': 78,
    'magenta': 50,
    'yellow': 67,
    'black': 80
}
```

```python
def plot_printer_toner_levels(printer_data, ax=None):
    ax.grid(axis='y')
    ax.set_facecolor('azure')
    colors = ['cyan', 'magenta', 'yellow', 'black']
    label_colors = ['k', 'k', 'k', 'w']
    
    amounts = [printer_data[c] for c in colors]
    top = [100 - amount for amount in amounts]
    p = ax.bar(x=colors, height=amounts, zorder=2, color=colors, edgecolor='#666')
    ax.bar(x=colors, bottom=amounts, height=top, zorder=2, color='lightgray', alpha=0.6, linewidth=1)
    labels = ax.bar_label(p, label_type='center', fontsize=30, color='w')
    for label, label_color in zip(labels, label_colors):
        label.set_color(label_color)
    ax.set_title(printer_data['name'] + ' - ' + printer_data['ip_address'])
```

```python
fig, ax = plt.subplots()

plot_printer_toner_levels(printer_data, ax = ax)

plt.show()
```
