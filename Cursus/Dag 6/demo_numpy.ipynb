```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
getallen = [23,45,67,89]
print(getallen)

maal_honderd = []
for getal in getallen:
    maal_honderd.append(getal * 100)

# list comprehension
maal_honderd = [getal * 100 for getal in getallen]

print(maal_honderd)
```

```python
a = np.array(getallen)
print(a)

a_honderd = a * 100
print(a_honderd)
```

```python
getallen = []
for getal in range(1000):
    getallen.append(random.randint(1, 100))

len(getallen)
```
